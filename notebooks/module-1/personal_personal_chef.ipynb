```python
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from langchain.tools import tool
from langgraph.checkpoint.memory import InMemorySaver
from pprint import pprint
from typing import Dict, Any
from tavily import TavilyClient
```

```python
load_dotenv()
```

```python
system_prompt = """
Be brief. You are cullinar Chef, with access to web. Based on ingredients i will tell you, find me receipe in internet based on given ingredients. Than ask me if it's ok to buy 
some more needed ingredients, and than propose meal with ingredients i have and the ones i decided to buy, and than present it to me with given formula. 
Ingredients: listed ingredients with volumes
Preparation phase: listed preparation phase
Cooking phase: listed cooking steps.
"""
```

```python
tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:
    """Search the web for receipes"""
    return tavily_client.search(query)
```

```python
agent = create_agent(
    model="gpt-5-nano",
    tools=[web_search],
    system_prompt=system_prompt,
    checkpointer=InMemorySaver()
)
```

```python
config = {"configurable": {"thread_id": "1"}}
prompt = HumanMessage(content="Hey, make me a meal with paprica, chicken breasts and chow main.")
```

```python
response = agent.invoke(
    {"messages": [prompt]},
    config
)
```
